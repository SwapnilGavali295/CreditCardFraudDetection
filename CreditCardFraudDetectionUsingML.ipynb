# Credit Card Fraud Detection using ML
The dataset provided contains credit card transactions with 31 features and a class label. These features describe different aspects of each transaction, such as time, amount, and other anonymized variables obtained through principal component analysis (PCA). The class label indicates whether a transaction is fraudulent (class 1) or not (class 0).

The "Time" feature represents the time elapsed in seconds between the transaction and the first transaction in the dataset. The features V1 to V28 are the anonymized variables resulting from the PCA transformation, representing various transaction characteristics.

The "Amount" feature denotes the transaction amount in USD, and the final feature is the "Class" label, determining whether the transaction is fraudulent or not.

In summary, this dataset is used to train machine learning models that aim to detect fraudulent transactions in real-time. By learning patterns from the provided features, the models can identify potential instances of fraud in future transactions.

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
credit_card_data = pd.read_csv('creditcard.csv')
credit_card_data.head()
```

```python
credit_card_data.sample()
```

```python
# dataset informations
credit_card_data.info()
```

```python
# checking the number of missing values in each column
credit_card_data.isnull().sum()
```

```python
# distribution of legit transactions & fraudulent transactions
credit_card_data['Class'].value_counts()
```

```python
legit = credit_card_data[credit_card_data.Class==0]
fraud = credit_card_data[credit_card_data['Class']==1]
```

```python
fraud['Class']
```

```python
# statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare the values for both transactions
credit_card_data.groupby('Class').mean()
```

```python
legit_sample = legit.sample(n=492)
```

```python
new_df = pd.concat([legit_sample,fraud],axis=0)
```

```python
new_df
```

```python
new_df['Class'].value_counts()
```

```python
new_df.groupby('Class').mean()
```

```python
X = new_df.drop(columns='Class', axis=1)
Y = new_df['Class']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

# Training model using Logistic Regression


```python
model=LogisticRegression()
```

```python
# training the Logistic Regression Model with Training Data
model.fit(X_train, Y_train)
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print('Accuracy score on Test Data : ', test_data_accuracy)
```
